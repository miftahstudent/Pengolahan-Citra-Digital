1. Import Dependency

```python
import imageio #Untuk membaca gambar dari file.
import matplotlib.pyplot as plt #Untuk menampilkan gambar.
import numpy as np #Untuk manipulasi array yang efisien.
import random #Untuk menghasilkan angka acak.
```

2. Load Image

```python
image_original = imageio.imread("gambar4.jpg")
```

Membaca gambar "gambar4.jpg" ke dalam variabel image_original.
Proses ini mempersiapkan gambar untuk manipulasi lebih lanjut.

3. Show Image

```python
# Display the loaded image
plt.imshow(image_original)
plt.title("Loaded Image")
plt.show()
```

Menampilkan gambar yang sudah dimuat.

4. To Grayscale

```python
gray_image = np.zeros_like(image_original)
for row in range(image_original.shape[0]):
    for col in range(image_original.shape[1]):
        pixel = image_original[row, col]
        avg_intensity = sum(pixel[:3]) // 3  # Calculate average of RGB values
        gray_image[row, col] = (avg_intensity,) * 3

plt.imshow(gray_image)
plt.title("Grayscale Image")
plt.show()
```

Mengonversi gambar ke skala abu-abu dengan menghitung rata-rata intensitas piksel.
Skala abu-abu adalah format yang lebih sederhana untuk manipulasi gambar.

5. Noise Gaussian

```python
gaussian_noise_image = np.copy(gray_image)
for row in range(image_original.shape[0]):
    for col in range(image_original.shape[1]):
        intensity = gray_image[row, col, 0]
        if random.randint(0, 99) < 20:  # 20% probability
            noise = random.randint(-128, 127)
            noisy_value = np.clip(intensity + noise, 0, 255)
            gaussian_noise_image[row, col] = (noisy_value,) * 3

plt.imshow(gaussian_noise_image)
plt.title("Gaussian Noise")
plt.show()
```

Menambahkan Gaussian noise dengan peluang 20% menggunakan nilai acak dari -128 hingga 127.
Gaussian noise sering digunakan untuk mensimulasikan gangguan pada gambar.

6. Noise Speckle

```python
speckle_noise_image = np.copy(gray_image)
for row in range(image_original.shape[0]):
    for col in range(image_original.shape[1]):
        if random.randint(0, 99) < 20:  # 20% probability
            speckle_noise_image[row, col] = (0,) * 3  # Black pixel for speckle noise

plt.imshow(speckle_noise_image)
plt.title("Speckle Noise")
plt.show()
```

Menambahkan speckle noise dengan mengganti piksel menjadi hitam (nilai 0) secara acak dengan peluang 20%.
Speckle noise biasanya muncul pada citra fotografi berbasis radar.

7. Noise Salt and Pepper

```python
salt_pepper_noise_image = np.copy(gray_image)
for row in range(image_original.shape[0]):
    for col in range(image_original.shape[1]):
        if random.randint(0, 99) < 20:  # 20% probability
            salt_pepper_noise_image[row, col] = (255,) * 3  # White pixel for salt noise

plt.imshow(salt_pepper_noise_image)
plt.title("Salt and Pepper Noise")
plt.show()
```

Menambahkan salt noise dengan mengganti piksel menjadi putih (nilai 255) secara acak dengan peluang 20%.
Noise ini mensimulasikan gangguan pada data gambar digital.

Kesimpulan

Proses ini terdiri dari pembacaan gambar, konversi skala abu-abu, dan penambahan berbagai jenis noise (Gaussian, speckle, dan salt-and-pepper).
Teknik manipulasi gambar ini sangat berguna untuk melatih algoritma pemrosesan gambar dalam kondisi berbeda.
